```python
def basePlot(ax):
  # plot 함수를 그릴때, figure에서 원하는 위치를 지정하기 위해 ax 속성을 사용
  mower.loc[mower.Ownership == 'Owner'].plot(x='Income', y='Lot_Size',style='o',
                                             markerfacecolor='red', markeredgecolor='red', ax = ax)
  mower.loc[mower.Ownership == 'Nonowner'].plot(x='Income', y='Lot_Size',style='o',
                                             markerfacecolor='none', markeredgecolor='blue', ax = ax)
  ax.legend(["Owner","Nonowner"], loc="upper left") # 범례 위치 조정 코드 추가
  ax.set_xlim(20, 120) # Income 나타낼 범위
  ax.set_ylim(13, 25) # Lot_Size 나타낼 범위
  ax.set_xlabel('Income')
  ax.set_ylabel('Lot Size')
  # return ax

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(11,4))
# print(fig)
# print(axes)
# print(axes[0])

ax = basePlot(axes[0]) # 첫 번째 위치에 그림을 그리겠다.
# print(type(ax))
y0, y1, y2 = 21.3, 19.8, 16.4
x0, x1, x2 = 55, 58, 95
axes[0].plot((20, x0), (y0,y0), color='grey') # return ax 없을 때
axes[0].plot((20, 120), (y1,y1), color='grey')
axes[0].plot((x1, x2), (y2,y2), color='grey')
axes[0].plot((x0, x0), (25,y1), color='grey')
axes[0].plot((x1, x1), (y1,13), color='grey')
axes[0].plot((x2, x2), (y1,13), color='grey')

ax = basePlot(axes[1])
plt.plot((40, 95), (25,13), color='grey') # plt는 항상 제일 마지막 그림에 추가됨

plt.tight_layout() # 간격 조정
plt.show()
```
