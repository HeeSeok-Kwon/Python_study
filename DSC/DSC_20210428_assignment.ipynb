```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 
```

```python
get_job = pd.read_csv("한국기술교육대학교 취업률 및 취업분야 현황_20200918.csv")
get_job
```

```python
get_job.info()
```

```python
get_job = get_job.set_index("졸업년도")
get_job
```

```python
import matplotlib.ticker as ticker
ax=plt.axes()

get_job['취업률'].plot(marker="o", color=["purple"])
plt.title("한국기술교육대학교 연도별 취업률 현황") # 제목 지정
plt.ylabel("취업률") # y축 이름 지정
plt.ylim(50, 100) # y축 범위 지정
ax.xaxis.set_major_locator(ticker.MultipleLocator(1)) # x축 간격 조정
```
