```python
pname = fpop.PNU_NAME
pname[1:100]
```

```python
i = 0
P1 = [] # 도단위
P2 = [] # 시단위
P3 = [] # 읍면동 단위
while i < len(fpop):
  temp = pname[i].split()
  P1.insert(i, temp[0]) # 제주특별자치도 추출
  P2.insert(i, temp[1]) # 제주시 또는 서귀포시 추출
  P3.insert(i, temp[2]) # 읍면동 추출
  i += 1
print(type(P1), len(P1), len(fpop))
```

```python
fpop['P1'] = P1
fpop['P2'] = P2
fpop['P3'] = P3
fpop.columns
```

```python
group_fpop = fpop.groupby(['P3']).sum()
len(group_fpop)
group_fpop
```

```python
type(group_fpop)
```

```python
group_fpop.iloc[2]['M10']
```

```python
i = 0
TOTAL = []
WOMAN = []
MAN = []
while i < len(group_fpop):
  each_man = 0
  for j in ('M10', 'M20', 'M30', 'M40', 'M50', 'M60'):
    each_man += group_fpop.iloc[i][j]
  MAN.insert(i,each_man)
  each_woman = 0
  for r in ('W10', 'W20', 'W30', 'W40', 'W50', 'W60'):
    each_woman += group_fpop.iloc[i][j]
  WOMAN.insert(i, each_woman)
  each_total = each_man + each_woman
  TOTAL.insert(i, each_total)
  i += 1

group_fpop['MAN'] = MAN
group_fpop['WOMAN'] = WOMAN
group_fpop['TOTAL'] = TOTAL
group_fpop
```

```python
group_fpop_new = group_fpop[['M10','M20', 'M30','M40','M50','M60','W10','W20','W30','W40', 'W50', 'W60', 'MAN', 'WOMAN', 'TOTAL' ]]
group_fpop_new.to_csv('group_fpop.csv', encoding='CP949')
```
