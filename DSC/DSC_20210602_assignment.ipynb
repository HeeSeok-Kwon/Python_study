```python
import math
import numpy as np
```

```python
# 비트코인 수정주가 차분
btc_r = []
btc_r.append(0) # 0번째 인덱스는 안쓰니까 임의로 아무 숫자 넣어줌
buf = btc.Adj_Close # buf 변수에 할당
size = len(buf) # buf 길이 변수 선언
for i in range(1, size): # 수익률 계산
  res = np.log(buf[i]/buf[i-1])
  btc_r.append(res)

btc_r = pd.DataFrame(btc_r) 
btc_r.head()
```

```python
# 이러리움 수정주가 차분
eth_r = []
eth_r.append(0) # 0번째 인덱스는 안쓰니까 임의로 아무 숫자 넣어줌
buf = eth.Adj_Close # buf 변수에 할당
size = len(buf) # buf 길이 변수 선언
for i in range(1, size): # 수익률 계산
  res = np.log(buf[i]/buf[i-1])
  eth_r.append(res)

eth_r = pd.DataFrame(eth_r) 
eth_r.head()
```

```python
btc["BTC_r"] = btc_r
btc.head()
```

```python
eth["ETH_r"] = eth_r
eth.head()
```

```python
# 수익률 시각화
plt.figure(figsize=(24,12))
plt.plot(btc.Date[1:], btc.BTC_r[1:])
plt.xticks(rotation=90)
plt.title("일별 비트코인 수익률 추이", fontsize = 20)
plt.xlabel("날짜")
plt.ylabel("비트코인 수익률")
ax=plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(3)) # 3일 간격으로 보여줌
```

```python
# 수익률 시각화
plt.figure(figsize=(24,12))
plt.plot(eth.Date[1:], eth.ETH_r[1:])
plt.xticks(rotation=90)
plt.title("일별 이더리움 수익률 추이", fontsize = 20)
plt.xlabel("날짜")
plt.ylabel("비트코인 수익률")
ax=plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(3)) # 3일 간격으로 보여줌
```

```python
# 시계열 산점도
plt.figure(figsize=(24,12))
plt.scatter(btc.BTC_r[1:], eth.ETH_r[1:])
plt.xticks(rotation=90)
plt.title("시계열 산점도", fontsize = 20)
plt.xlabel("비트코인 수익률")
plt.ylabel("이더리움 수익률")
```

```python
from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression
```

```python
crypto_data = {"btc":btc.BTC_r[1:], "eth":eth.ETH_r[1:]}
crypto_data = pd.DataFrame(crypto_data)
crypto_data
```

```python
model = ols('eth ~ btc',data=crypto_data).fit() 
```

```python
model.summary()
```

```python
print(model.params.Intercept) # 절편
print(model.params.btc)  # 기울기
```

```python
# 시계열 회귀분석 시각화
plt.figure(figsize=(24,12))
plt.scatter(crypto_data.btc, crypto_data.eth)
plt.plot(crypto_data.btc, model.fittedvalues, color="red")
plt.title("시계열 회귀분석", fontsize = 20)
plt.xlabel("비트코인 수익률")
plt.ylabel("이더리움 수익률")
```
