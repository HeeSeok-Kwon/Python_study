```python
# 연습문제 12
X = np.array([1,2,3,4])
p = np.array([1/8,3/8,3/8,1/8])
m = 0
s = 0
for i in range(4):
    m += (p[i]*X[i])
    s += (p[i]*X[i]**2)
s -= m**2
print("Original mean is",m)
print("Original std is",np.sqrt(s))

m = 2*m+1
s = 4*s
print("Changed mean is",m)
print("Changed std is",np.sqrt(s))

Y = 2*X+1
m = 0
s = 0
for i in range(4):
    m += (p[i]*Y[i])
    s += (p[i]*Y[i]**2)
s -= m**2

print("Changed mean is",m)
print("Changed std is",np.sqrt(s))
```

```python
# Ch8 Programming Practice
import numpy as np
import matplotlib.pyplot as plt

# Define the probability density function for the normal distribution
def f(x,m,sigma):
    coef = 1/(np.sqrt(2*np.pi)*sigma)
    return coef*np.exp(-(x-m)**2/(2*sigma**2))

# Function value and plot
x = np.linspace(-10,10,101)
# print(x)
m = 0
for sigma in [0.5,1,2,4,8]:
    fx = f(x,m,sigma)
    s = str(sigma)
    plt.plot(x,fx,label=r'$\sigma$='+s)
plt.legend()
plt.show()
```
