```python
#Ex 2-6
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 2, 301)
y1 = 3**x
y2 = 3**(x-1) + 2

graph1, = plt.plot(x, y1)
graph2, = plt.plot(x, y2)
plt.legend(handles=(graph1, graph2), labels=(r'$y=3^x$', r'$y=3^{x-1}+2$'))
plt.show()

x = np.linspace(-1, 2, 301)
y1 = 2**(-x)
y2 = -2**(-x-2)

graph1, = plt.plot(x, y1)
graph2, = plt.plot(x, y2)
plt.legend(handles=(graph1, graph2), labels=(r'$y=2^{-x}$', r'$y=-2^{-x-2}$'))
plt.show()
```

```python
# 9
# 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 601)
y = (1/2)**(x+1) - 3
plt.plot(x, y)
plt.show()
print(":",max(y),", :",min(y))

#
# f(x)=(0.5)^(x+1)-3
def f(x):
	return (0.5)**(x+1)-3

# Plot the graph of y=f(x)
x = np.linspace(-2,4,601)
fx = f(x)
plt.plot(x,fx)
plt.show()
print("Min of f : ", f(4))
print("Max of f : ", f(-2))
```

```python
# 10
# 
import numpy as np
import matplotlib.pyplot as plt
import math

x1 = np.linspace(0, 1000, 10000)
base = 3
x2 = []
y = []
for i in range(0, 9999):
    x2.append(math.log(x1[i+1], base))
    y.append(x2[i]**2 - 4*x2[i] + 3)
plt.plot(x2, y)
plt.show()
print(": , :",round(min(y),1))

# 
# f(x)=(log_3(x))^2-log_3(x^4)+3
def f(x):
	return (np.log(x)/np.log(3))**2-np.log(x**4)/np.log(3)+3

# Plot the graph of y=f(x)
x = np.linspace(0.01,81,1001)
fx = f(x)
plt.plot(x,fx)
plt.show()
```
