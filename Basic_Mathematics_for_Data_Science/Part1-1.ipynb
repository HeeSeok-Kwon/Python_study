```python
#연습문제 4번
def f(x):
    buf = []
    for i in x:
        if i <= -1: buf.append(i+2)
        else : buf.append(i**2)
    return buf
def g(x):
    return x + abs(x)

x = np.linspace(-4, 3, 401)
graph1, = plt.plot(x, f(x))
graph2, = plt.plot(x, g(x))
plt.legend(handles=(graph1, graph2), labels=('a','b'))
plt.show()
```

```python
#연습문제 8번
def even_function(x):
    return x**2
def odd_function(x):
    return x**3

x = np.linspace(-3, 3, 601)
graph1, = plt.plot(x, even_function(x))
graph2, = plt.plot(x, odd_function(x))
plt.legend(handles=(graph1, graph2), labels=('y=x^2','y=x^3'))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

```python
#연습문제 12
def a(x):
    return ((x-3)**2 - 1)/2
def b(x):
    return (1-x)/(2*x+1)

x = np.linspace(0,3,601)
graph1, = plt.plot(x, a(x))
graph2, = plt.plot(x, b(x))
plt.legend(handles=(graph1, graph2), labels=('y=a(x)','y=b(x)'))
plt.xlim(-1,3)
plt.ylim(-1,3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    if x >= -1 and x < 1:
        return -x**2 + 2
    elif x >= 1 and x <= 3:
        return x**2 -4*x +3
    elif x > 3 and x <= 5:
        return -x**2 + 8*x - 14

x = np.linspace(-1, 5, 601)
fx = np.zeros_like(x) # x와 사이즈가 같고 0으로 초기화된 배열

for i in range(0, 601):
    fx[i] = f(x[i])
    
plt.plot(x, fx, '.')
plt.show()
```
