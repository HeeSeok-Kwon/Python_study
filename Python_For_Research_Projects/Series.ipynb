제 4절 시리즈 구조

```python
import pandas as pd
import numpy as np
```

```python
se = pd.Series([10, 40, 30, 20, 50])
se
```

```python
se.values
```

```python
se.index
```

```python
se.dtypes
```

```python
se = list(map(float, se)) # Series element to float
se
```

```python
se = list(map(int, se)) # Series element to integer
se
```

```python
se = pd.Series(se) # Series type casting
se
```

```python
seCpy1 = pd.Series(se.values, index=['kim', 'lee', 'park', 'kang', 'no'])
seCpy1
```

```python
seTmp1 = pd.Series(np.arange(5.), index=['a', 'b', 'c', 'd', 'e'])
seTmp1
```

```python
birth = {'kim':1990, 'lee':1992, 'park':1990, 'kang':1993, 'no':1983}
seCpy2 = pd.Series(birth) # dictionary to Series
seCpy2
```

```python
seCpy2.name = 'Birth Year'
seCpy2.index.name = 'Student'
seCpy2
```

```python
seCpy2.index = [1, 2, 3, 4, 5]
seCpy2
```
