```python
# 기상청 사이트에서 전국 날씨 정보 크롤링

import requests # 웹페이지 접근 요청 
from bs4 import BeautifulSoup # 크롤링
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)
driver.get("https://www.weather.go.kr/w/index.do")
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
# print(html)
```

```python
weather = soup.find("div",{"id","container"}).find("div",{"class":"lay-460"})
# print(weather)
```

```python
wdict = {'location':[], 'weather':[], 'temperature':[],'precipitation':[]}
print(wdict)
```

```python
del wdict['location']
print(wdict)
```

```python
map1 = weather.find("div",{"id":"weather2"})
# print(map1)
```

```python
location = map1.find_all("dt")
# print(location)
for i in range(0, len(location)):
  # print(location[i].get_text())
  wdict['location'].append(location[i].get_text())
print(wdict['location'])
```

```python
wth = map1.find_all("img")
for w in wth:
  # print(w['alt'])
  wdict['weather'].append(w['alt'])
# print(wdict)
```

```python
temperature = map1.find_all("span")
for t in temperature:
  wdict['temperature'].append(t.get_text())
# print(wdict)
```

```python
map2 = weather.find("div",{"id":"rainfall"})
# print(map2)
```

```python
precipitation = map2.find_all("dd")
# print(precipitation)
for p in precipitation:
  wdict['precipitation'].append(p.get_text())
print(wdict)
```
