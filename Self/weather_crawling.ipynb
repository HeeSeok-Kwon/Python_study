```python
# 기상청 사이트에서 전국 날씨 정보 크롤링

import requests # 웹페이지 접근 요청 
from bs4 import BeautifulSoup # 크롤링
import pandas as pd # Type을 DataFrame으로 바꿔주기 위한 코드

url = 'https://www.weather.go.kr/m/nation/current.jsp?ele=1'
wdict = {'location':[], 'weather':[], 'temperature':[],'precipitation':[]} # 크롤링한 데이터를 담기 위한 딕셔너리
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    location = soup.select('#weather2 > dl > dt') # 장소 정보 추출
    # print(location)
    for i in range(0, len(location)): 
      weather = soup.select('#weather2 > dl > dd > a > img')[i]['alt'] # 날씨 정보 추출
      wdict['location'].append(location[i].get_text())
      wdict['weather'].append(weather)
      # print(location[i].get_text(),":",weather)
else : 
    print(response.status_code)
```
